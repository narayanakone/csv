Based on last 5 year statistics, an app based food delivery company declared its statistics at Chennai, saying the average delivery time to any part of Chennai is 45 min, with a standard deviation of 20 min. Compute the following,
a) What proportion of delivery were done less than 30 min?
b) What proportion of delivery were done more than 100 min?
c) What proportion of delivery were done between 45min to 60min?

```python
stats.norm.cdf(30,45,20)
```

```python
stats.norm.cdf((30-45)/20)
```

```python
stats.norm.cdf(60,45,20) -stats.norm.cdf(45,45,20)
```
